Skip numbers divisible by 5 & stop if number > 50

```python
numbers = [10, 12, 25, 30, 40, 55, 18]

for n in numbers:
    if n > 50:
        break
    if n % 5 == 0:
        continue
    print(n)
```

Reverse every alternate word in a string

```python
s = input("Enter a sentence: ")

words = s.split()
new_words = []

for i in range(len(words)):
    if i % 2 == 1:
        new_words.append(words[i][::-1])
    else:
        new_words.append(words[i])

result = " ".join(new_words)
print(result)
```

Find repeated words and store them in a dictionary

```python
words = ["apple", "banana", "apple", "orange", "banana", "banana"]

freq = {}

for w in words:
    if w in freq:
        freq[w] = freq[w] + 1
    else:
        freq[w] = 1

repeat = {}

for k in freq:
    if freq[k] > 1:
        repeat[k] = freq[k]

print(repeat)
```

Book availability program (with input validation)

```python
books = {
    'Book1': 5,
    'Book2': 6,
    'Book3': 10
}

book_name = input("Enter book name: ")

if book_name not in books:
    print("Unavailable")
else:
    while True:
        copies = input("How many copies do you want? ")

        if copies.isdigit():
            copies = int(copies)
            break
        else:
            print("Please enter a valid number.")

    available = books[book_name]

    if available >= copies:
        print("Available")
    elif available < copies:
        print("Partially Available")
```

Store frequency of each word in dictionary

```python
words = ["This", "is", "good", "is"]

freq = {}

for w in words:
    w2 = w.lower()
    if w2 in freq:
        freq[w2] = freq[w2] + 1
    else:
        freq[w2] = 1

print(freq)
```

Function to return the largest word

```python
def largest_word(sentence):
    words = sentence.split()
    biggest = words[0]

    for w in words:
        if len(w) > len(biggest):
            biggest = w

    return biggest

print(largest_word("Python programming is awesome"))
```

Function to convert to title case

```python
def to_title_case(sentence):
    words = sentence.split()
    new = []

    for w in words:
        new.append(w.capitalize())

    return " ".join(new)

print(to_title_case("hello world from python"))
```
